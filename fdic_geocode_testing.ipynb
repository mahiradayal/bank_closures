```python
latitude = []
longitude = []

def geocode_address(loc):
    gmaps = googlemaps.Client(key='AIzaSyCyKn8jrgRCFeAPvBxt8aFOi_vPFWKFwHs')
    geocode_result = gmaps.geocode(loc)
    if len(geocode_result) > 0: 
        lat = geocode_result[0]["geometry"]["location"]["lat"]
        lon = geocode_result[0]["geometry"]["location"]["lng"]
        latitude.append(lat)
        longitude.append(lon)
    else: 
        lat = '00'
        lon = '00'
        latitude.append(lat)
        longitude.append(lon)
```

```python
for ADDRESSES in closure_metros.ADDRESS_FULL:
    geocode_address(ADDRESSES) 
```
